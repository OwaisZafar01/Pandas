# Merging, Joining, Concatenation


```python
import numpy as np
import pandas as pd

courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
reg1 = pd.read_csv('reg-month1.csv')
reg2 = pd.read_csv('reg-month2.csv')
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
```

## Concat

```python
# pd.concat

regs = pd.concat([reg1,reg2], axis=0, ignore_index=True)
regs
```

```python
pd.concat([reg1,reg2], axis=1)
```

```python
# multiindex dataframe

multi = pd.concat([reg1, reg2], keys=['Nov','Dec'])
multi
#multi.loc['Nov']
#multi.loc[('Nov',5)]
```

## Merge

```python
# inner joins sirf wahin records visible honge jo dono tables me honge

students.merge(regs, how='inner',on='student_id')
```

```python
# left join  left table ka content aur dono me common records ko print karega
courses.merge(regs, how='left',on='course_id')
```

```python
# right join  right table ka content aur dono me common records ko print karega

temp = pd.DataFrame({
        'student_id': [26,27,28],
        'name':['Owais','Huzaifa','Zafar'],
        'partner':[28,26,17]})

students = pd.concat([students,temp], ignore_index=True)
students
```

```python
students.merge(regs, how='right',on = 'student_id')
```

```python
# full join dono tables ka complete record print karega

students.merge(regs, how='outer',on='student_id').tail(5)
```

### Joins Exercise

```python
# find total revenue generated

total = regs.merge(courses, how='inner',on='course_id')['price'].sum()
print(total)
```

```python
# find month by month revenue

temp = multi.reset_index()
temp = temp.merge(courses, how='inner', on='course_id').groupby('level_0')['price'].sum()
temp
```

```python
# print the registration table
# cols name course price

regs.merge(students, on='student_id').merge(courses, on='course_id')[['name','course_name','price']]
```

```python
# find students who enrolled in both the months

com_stds = np.intersect1d(reg1['student_id'],reg2['student_id'])
students[students['student_id']. isin(com_stds)]
```

```python
# find course that got no enrollment

c = np.setdiff1d(courses['course_id'], regs['course_id'])
courses[courses['course_id'].isin(c)]
```

```python
# find students who did not enroll into any courses

c = np.setdiff1d(students['student_id'], regs['student_id'])
students[students['student_id'].isin(c)]
```

```python
# print student name partner name for all enrolled students
# self join
students.merge(students, how='inner',left_on = 'partner', right_on = 'student_id')[['name_x','name_y']]
```

```python
# find top 3 students who did most number of enrollments
regs.merge(students, on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# find top 3 students who spant most amount of money on courses
regs.merge(students, on='student_id').merge(courses, on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# find top 3 stadiums with highest sixes/match ratio

temp = pd.merge(deliveries,matches, how='inner', left_on = 'match_id', right_on = 'id')
temp = temp[temp['batsman_runs']==6]
sixes = temp.groupby('venue')['venue'].count()
total = matches['venue'].value_counts()
res = sixes / total
res.sort_values(ascending=False).reset_index().head(3)
```

```python
# find orange cap holder of all seasons
temp = pd.merge(deliveries,matches, how='inner', left_on = 'match_id', right_on = 'id')
temp.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs', ascending=False).drop_duplicates(subset=['season'], keep='first').sort_values('season')
```
