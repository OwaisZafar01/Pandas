# MultiIndex

```python
# multi index object
# pd.MultiIndex.from_tuples()
index_val = [('cse', 2019),('cse', 2020),('cse', 2021),('cse', 2022), ('ece', 2023),('ece', 2024),('ece', 2025),('ece', 2026)]
multiindex = pd.MultiIndex.from_tuples(index_val)
print(multiindex.levels[0])
print(multiindex.levels[1])
print(multiindex)
print('\n')
# pd.MultiIndex.from_product()
print(pd.MultiIndex.from_product([['cse','ece'],[2019, 2020,2021,2022,2023,2024,2025,2026]]))
```

```python
# creation a multiindex Series

s = pd.Series([1,2,3,4,5,6,7,8], index = multiindex)
s
```

```python
# Fetch items from a series

print(s['cse'])
print(s['cse',2022])
```

```python
# unstack

unstack = s.unstack()
unstack
```

```python
# stack
unstack.stack()
```

## Melt

```python
# melt wide to long
df = pd.DataFrame({'Maths':[120]})
df.melt()
```

```python
df = pd.DataFrame({'Maths':[120],'English':[150]})
df.melt(var_name='Subjects', value_name = 'No.of Students')
```

```python
df = pd.DataFrame({'Subject':['Maths','English','Physics'],'2020':[100,200,300],'2021':[10,20,30],'2022':[1,2,3]})
df.melt(id_vars=['Subject'])
```

```python
death = pd.read_csv('time_series_covid19_deaths_global.csv')
confirm = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
death.head(3)
```

```python
confirm.head(3)
```

```python
death = death.melt(id_vars = ['Province/State','Country/Region','Lat','Long'], var_name = 'Date', value_name='Confirm Death')
confirm = confirm.melt(id_vars = ['Province/State','Country/Region','Lat','Long'], var_name = 'Date', value_name='Confirm Cases')
```

```python
death.head(3)
```

```python
confirm.head(3)
```

```python
death.merge(confirm,
            how='inner',
            on=['Province/State','Country/Region','Lat','Long','Date'])[['Country/Region','Date','Confirm Death','Confirm Cases']].head(3)
```
